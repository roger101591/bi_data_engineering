```python
def create_first_df():    
    opt_ins_total_df = opt_ins_total(opt_ins)
    cust_orders_rolling_df = cust_orders_rolling()
    buckets_final_df = buckets_final(cust_orders_rolling_df)
    active_members_df = active_members(buckets_final_df)
    member_buckets_final_df = member_buckets_final(opt_ins_total_df, active_members_df, buckets_final_df)
    ever_redeemed_df2 = ever_redeemed(ever_redeemed_df)
    kpi_first_df2 = kpi_first_df(member_buckets_final_df, ever_redeemed_df2)
    return kpi_first_df2
```

```python
kpi_first_df = create_first_df()
```
